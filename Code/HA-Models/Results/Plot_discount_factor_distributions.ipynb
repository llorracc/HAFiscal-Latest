# Produce graph over discount factor distributions

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
import pandas as pd

plt.style.use('classic')
```

```python
CRRA = 1.5
R = 1.01
dfFileStr = 'DiscFacDistributions_CRRA_'+str(CRRA)+'_R_'+str(R)+'.txt'
dfFile = open(dfFileStr, 'r')
myDFs = [[],[],[]]
readStr = dfFile.readline().strip()
while readStr != '' :
    dictload = eval(readStr)
    edType = dictload['EducationGroup']
    theBetas = dictload['betaDistr']
    myDFs[edType] = theBetas
    readStr = dfFile.readline().strip()
dfFile.close()
```

```python
# Figure titles:
mytitles = ['Dropout', 'Highschool', 'College']
myYticks = np.linspace(0,1.2,7)
x_axis = range(1,8)
```

```python
fig = plt.figure()
for e in [0,1,2]:
    plt.plot(x_axis,myDFs[e][:], label=mytitles[e])
    
plt.yticks(ticks=myYticks)
plt.xlabel(xlabel='Type', fontsize=14)
plt.ylabel(ylabel='Discount factor', fontsize=14)
plt.legend(loc='lower right')
fig.set_facecolor(color="white")
plt.show()
```

```python
# Save fig:
fig.savefig('DFdistribution_CRRA_'+str(CRRA)+'.pdf', facecolor=fig.get_facecolor(), edgecolor='none')
```
